```python
def find_maximum(digits,str_number):
    maximum_value = 0
    for i in range(len(str_number)- digits+1):
        digit_number = str_number[i:i+digits]
        multiply_value = 1
        for j in digit_number:
            multiply_value *= int(j)
        if multiply_value > maximum_value:
            maximum_value = multiply_value
    return maximum_value
```
